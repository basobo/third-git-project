```python
df_day_hour2 = pd.pivot_table(data_df[['start_time_day_name', 'start_time_hour', 'duration_second','commodity','totalWeightSum']], index=['start_time_day_name', 'start_time_hour'], aggfunc='count')
```

```python
# Unstack below puts days on the x-axis and hour of day on the y-axis.
df_day_hour3 = df_day_hour2.unstack(level=0)
```

```python
# Create new axis tick labels for easy to read hours and days of the week.
morning_hours = []
for hour in range(1, 12):
    detailed_hour = str(hour) + "am"
    morning_hours.append(detailed_hour)
```

```python
afternoon_hours = []
for hour in range(1, 12):
    detailed_hour = str(hour) + "pm"
    afternoon_hours.append(detailed_hour)
```

```python
detailed_hours = ["12am"] + morning_hours + ["12pm"] + afternoon_hours
```

```python
day_short_names = ['Strawberries']
```

```python
sns.set_context("talk")
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(df_day_hour3, annot=True, linewidths=1, ax=ax, xticklabels=day_short_names, yticklabels=detailed_hours,cmap='winter_r')
ax.axes.set_title("Heatmap Counts by Day and Hour of Day", fontsize=24, y=1.01)
ax.set(xlabel='commodity', ylabel='Starting Hour');
```
